```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\00_portifolio\05_pdf_scrapper\data\processed\focus_annual.csv')
```

```python
print(f"Total registers: {len(df)}")
print(f"Period: {df['ref_date'].min()} a {df['ref_date'].max()}")
```

```python
years = [2023, 2024, 2025, 2026, 2027, 2028]

plt.figure(figsize=(12,6))

for year in years:
    df_year = df[(df['indicator']=='IPCA') & (df['year']==year)].copy()
    if not df_year.empty:
        df_year['ref_date'] = pd.to_datetime(df_year['ref_date'])
        plt.plot(df_year['ref_date'], df_year['value'], linestyle='-', label=str(year))

plt.title('IPCA expectation over time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
